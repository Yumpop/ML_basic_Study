# 회귀 

- 회귀는 여러 개의 독립변수(Feature)와 한개의 종속변수(target)간의 상관관계를 모델링하는 기법
- (종속변수) Y = (독립변수) X1 * (회귀계수)W1 +...+ Xn * Wn
- 머신러닝 회귀 예측은 주어진 독립변수와 종속변수의 데이터를 가지고 학습을 통해 최적의 회귀 함수를 찾아내는 것
- 최적의 회귀함수를 찾는다는 것은 전체 데이터의 잔차(오류 값, 예측모델과 실제 데이터 사이의 오류값) 합이 최소가 되는 모델을 만든다는 의미
- 

- 일반 선형회귀 : 예측값과 실제값의 RSS를 최소화 할수 있도록 회귀계수를 최적화하면서 규제를 적용하지 않음
- 릿지(Ridge) : 릿지회귀는 선형회귀에 L2규제를 추가한 회귀모델
- 라쏘(Lasso) : 라쏘회귀는 선형회귀에 L1규제를 적용한 방식
- 엘라스틱넷 : L1과 L2 규제를 함께 결합한 모델
- 로지스틱 회귀 : 분류에서 사용되는 선형모델

- RSS : 오류값의 제곱을 구해 다 더하는 방식
    - 제곱은 오류값이 음수가 될 수도 있기 때문, 또한 미분 등 계산이 편리하기 때문 
- RSS는 이제 회귀계수인 W1, W2를 중심변수로 하고 종속변수 Y, 독립변수 X들은 상수취급해서 결국 회귀 계수를 찾아가는 과정이다.
- 회귀에서 RSS는 비용이며 w변수로 구성되는 RSS를 비용함수라고 한다./ 비용함수 == 손실함수

image.png

- 비용 최소화 하기 
    - 대표적인 방법 - 경사하강법(Gradient Descent)
    - 점진적으로 반복적인 계산을 통해 W파라미터 값을 업데이트 하면서 오류값이 최소값이 되는 W값을 찾는 방법

- 오류가 작아지는 방향으로 W값을 보정하는 방법 ---> 미분
- 미분은 함수의 증가 또는 감소의 방향성을 나타낸다. 
- 만약 2차함수의 비용함수라고 하면 미분된 1차함수의 기울기가 더이상 감소하지 않는 지점일때 비용함수가 최소 지점으로 간주한다. 
- RSS의 편미분은 그냥 하면 됨 !
